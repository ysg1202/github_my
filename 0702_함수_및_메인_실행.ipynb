```python
def scan_front(x, y, obstacles, sensors):
    print("앞쪽 스캔 시작")

    if (x, y + 1) in obstacles:
        print("앞쪽 1칸 장애물 발견")
        sensors["앞"] = 1.0

    elif (x, y + 2) in obstacles:
        sensors["앞"] = 2.0
        print("앞쪽 2칸 장애물 발견")

    elif (x, y + 3) in obstacles:
        print("앞쪽 3칸 장애물 발견")
        sensors["앞"] = 3.0

    else:
        print("앞쪽 장애물 없음", "센서값 15.0유지")
        sensors["앞"] = 15.0

    print("앞쪽 스캔 종료")
```

```python
def update_sensor_data(position, obstacles):
    print("센서 업데이트 시작")

    x = position[0]
    y = position[1]
    print("위치 추출 완료")
    print(f"현재 x값:{x}, 현재 y값:{y}")

    sensors = init_sensors()
    sensors["앞"] = 15.0
    sensors["왼쪽"] = 15.0
    sensors["오른쪽"] = 15.0

    scan_front(x, y, obstacles, sensors)
    scan_left(x, y, obstacles, sensors)
    scan_right(x, y, obstacles, sensors)

    print("센서 데이터 업데이트 완료")
    print(sensors)

    return sensors
```

```python
safe_distance = 5

def detect_obstacles(sensors, safe_distance):
    print("장애물 감지 시작")

    print("sensors:", sensors)
    print("safe_distance:", safe_distance)

    obstacles = []

    for direction, distance in sensors.items():
        print(f"현재 방향:{direction}, 현재 거리{distance}")

        if distance < safe_distance:
            obstacles.append(direction)
            print("위험!", "목록에 추가됨")

        else:
            print("안전함")

    print("장애물 감지 완료")
    print(obstacles)

    return obstacles
```

```python
def decide_direction(obstacles):
    if len(obstacles) == 0:
        action = "직진"
        print("장애물 없음")

    elif "앞" in obstacles:
        print("앞에 장애물 있음, 좌우 확인")

        if "왼쪽" not in obstacles:
            action = "좌회전"
            print("왼쪽 안전")

        elif "오른쪽" not in obstacles:
            action = "우회전"
            print("오른쪽 안전")

        else:
            action = "정지"
            print("모든 방향 위험")

    else:
        action = "주의해서 직진"
        print("앞은 안전,", "옆만 위험")

    return action
```

```python
class SimpleCar:
    def __init__(self, name, x, y):
        self.name = name
        self.x = x
        self.y = y

    def execute_action(self, action):
        if action == "직진":
            self.move_forward()
        elif action == "좌회전":
            self.turn_left()
        elif action == "우회전":
            self.turn_right()
        elif action == "정지":
            self.stop()

    def move_forward(self):
        self.y += 1
        print(f"{self.name} -> 전진: y = {self.y}")

    def turn_left(self):
        self.x -= 1
        print(f"{self.name} -> 좌회전: x {self.x}")

    def turn_right(self):
        self.x += 1
        print(f"{self.name} -> 우회전: x {self.x}")

    def stop(self):
        print(f"{self.name} -> 정지: 위치 = ({self.x}, {self.y})")

    def run_simulation(self, obstacles):
        step = 0
        while step < 10:
            step += 1
            print(f"\n스텝{step}")
            position = (self.x, self.y)
            sensor_data = update_sensor_data(position, obstacles)

            detected = detect_obstacles(sensor_data, safe_distance)

            action = decide_direction(detected)

            self.execute_action(action)

            if action == "정지":
                print("차량이 정지했습니다. 시뮬레이션 종료")
                break
```
